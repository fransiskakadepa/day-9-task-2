```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Data Contoh
X = np.array ([1468007,1481814,1503485,1535900,1558494,1580517]).reshape(-1,1) #variabel independen X
Y = np.array ([471224,409145,388652,384773,367121,390870])#variabel independen Y

# Membuat Model regresi sederhana
Model = LinearRegression()

# Melatih Model pada data
Model.fit(X,Y)

# Menghitung prediksi Y berdasarkan Model
Y_pred = Model.predict(X)
# Menghitung koefisien determinasi (R-sguared)
r_sguared = r2_score(Y,Y_pred)

# Membuat plot scatter
plt.scatter(X,Y, label='Data Asli', color='blue')
plt.plot(Y_pred, color='red', linewidth=2, label='Garis regresi',)
plt.xlabel('penduduk (X)')
plt.ylabel('kemiskinan (Y)')
plt.legend()
plt.title('DATA TAHUN,JUMLAH PENDUDUK,JUMLAH KEMISKINAN')
plt.grid(True)

# Membuat prediksi menggunakan Model
Y_pred = Model.predict(X)

#Menampilakan koefision regresi
print('koefision Regresi(m):',Model.coef_[0])
print('intersep (b):',Model.intercept_)
print('koefision Determinasi (R-sguared) :',r_sguared)

# Plot data asli dan garis regresi
plt.scatter(X,Y, label='Data Asli')
plt.plot(X,Y_pred,color='red', linewidth=2, label='Garis Regresi')
plt.xlabel('skor tes(x)')
plt.ylabel('hasil produksi (Y)')
plt.legend()
plt.show()
```
